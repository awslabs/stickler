# JSON Schema Import/Export Example

This notebook demonstrates creating a StructuredModel from a JSON Schema.

```python
from stickler import StructuredModel
```

## Define JSON Schema

Create a JSON Schema with Stickler extensions for comparison configuration.

```python
product_schema = {
    "type": "object",
    "x-aws-stickler-model-name": "Product",
    "properties": {
        "name": {
            "type": "string"
        },
        "price": {
            "type": "number"
        },
        "in_stock": {
            "type": "boolean"
        }
    },
    "required": ["name", "price"]
}
```

## Create Model from Schema

```python
Product = StructuredModel.from_json_schema(product_schema)
```

## Create Instances and Compare

```python
ground_truth = Product(name="Laptop", price=999.99, in_stock=True)
prediction = Product(name="Laptop Pro", price=999.99, in_stock=True)

result = ground_truth.compare_with(prediction)
print(f"Overall Score: {result['overall_score']:.3f}")
print(f"Field Scores: {result['field_scores']}")
```

## Export Model Configuration

Export the Product model back to JSON Schema format.

```python
exported_schema = Product.to_json_schema()
print(exported_schema)
```

## Export to Stickler Config

Export the Product model to custom Stickler configuration format.

```python
exported_config = Product.to_stickler_config()
print(exported_config)
```

## Round-trip Example

Export and re-import to verify configuration is preserved.

```python
# Export
schema = Product.to_json_schema()

# Re-import
ReconstructedProduct = StructuredModel.from_json_schema(schema)

# Verify behavior is identical
p1 = Product(name="Laptop", price=999.99, in_stock=True)
p2 = ReconstructedProduct(name="Laptop", price=999.99, in_stock=True)

result1 = p1.compare_with(p1)
result2 = p2.compare_with(p2)

print(f"Original score: {result1['overall_score']}")
print(f"Reconstructed score: {result2['overall_score']}")
print(f"Match: {result1['overall_score'] == result2['overall_score']}")
```
